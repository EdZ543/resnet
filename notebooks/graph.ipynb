```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_path = "wandb_export_2024-05-02T11_53_52.977-04_00.csv"

df = pd.read_csv(file_path, index_col=0)
```

```python
df.head()
```

```python
layers = [20, 32, 44]
colours = ["y", "c", "g"]

fig = plt.figure(figsize=(10, 7))
plt.axis([0, 182, 0, 20])

for layer, colour in zip(layers, colours):
    train_error = df[f"ResNet-{layer} - train/error"]
    val_error = df[f"ResNet-{layer} - validation/error"]

    plt.plot(df.index, train_error * 100, linestyle="--", color=colour)
    plt.plot(df.index, val_error * 100, label=f"ResNet-{layer}", color=colour)

plt.legend()
plt.xlabel("epochs")
plt.ylabel("error (%)")
plt.axhline(10, color="black", linestyle="--")
plt.axhline(5, color="black", linestyle="--")
```
